## Part B: Data Cleaning and Preprocessing (5 Marks) 

* Clean the data if required: 

```python
# 1.If there are missing values in critical columns (e.g., City, Sales), decide whether to drop them or impute them (e.g., using mean or median). 
#there is no missing values but i will assume that there are in the Sales coulumns
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
#2.Check for outliers or invalid data entries (e.g., negative Sales or unrealistic Discount rates) and handle them appropriately
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Keep only values within the bounds
    df_clean = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_clean

# Apply it to your needed columns
columns_to_check = ['Discount', 'Profit', 'Quantity', 'Row.ID', 'Sales', 'Shipping.Cost']

for col in columns_to_check:
    df = remove_outliers_iqr(df, col)
```

* Prepare the data for visualization: 

```python
#1.	Filter and structure the data for upcoming visualizations

# Convert the columns to datetime first (if not already)
df['Order.Date'] = pd.to_datetime(df['Order.Date'])
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])

# Remove the time part by keeping only the date
df['Order.Date'] = df['Order.Date'].dt.date
df['Ship.Date'] = df['Ship.Date'].dt.date

# Make sure they are datetime and remove time visually
df['Order.Date'] = pd.to_datetime(df['Order.Date']).dt.floor('D')
df['Ship.Date'] = pd.to_datetime(df['Ship.Date']).dt.floor('D')
```

* Heatmap: Product Categories in Top 10 Cities: (20 Marks) 

```python
#1.	Identify the top 10 cities based on the highest number of orders or total sales. 
# Use the original DataFrame, not the New York filtered one
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
print(city_sales)
```

```python
#2.	Generate a heatmap showing the total sales (or number of orders) for each Product Category 
#(e.g., Furniture, Office Supplies, Technology) across these top 10 cities. 
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Work on the original df
# Group by 'City' and 'Category', and sum the 'Sales'
city_category_sales = df.groupby(['City', 'Category'])['Sales'].sum().reset_index()

# Step 2: Get the Top 10 Cities by total Sales
top_10_cities = city_category_sales.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10).index

# Step 3: Filter the data only for Top 10 Cities
filtered_data = city_category_sales[city_category_sales['City'].isin(top_10_cities)]

# Step 4: Pivot the table: Rows = City, Columns = Category, Values = Sales
pivot_table = filtered_data.pivot(index='City', columns='Category', values='Sales')

# Step 5: Create the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='RdBu')

# Step 6: Add title and axis labels
plt.title('Total Sales for Each Product Category Across Top 10 Cities', fontsize=16)
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('City', fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#3.	Ensure clear labeling of the axes and color scale so the heatmap is easy to interpret. 
plt.figure(figsize=(12,8))

# Create the heatmap
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='RdBu', 
            cbar_kws={'label': 'Total Sales'},  # Color bar label
            linewidths=0.5, linecolor='gray',   # Adds grid lines for better readability
            cbar=True)  # Display the color bar

# Step 6: Add title and axis labels with appropriate font size
plt.title('Total Sales for Each Product Category Across Top 10 Cities', fontsize=16)
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('City', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Keep y-axis labels horizontal (optional)
plt.yticks(rotation=0)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

* Bar Plot: Average Profit Across Cities for a Specific Time Period: (20 
Marks) 


```python
#1.	Filter the dataset to focus on a particular time period (e.g., the month of December or Q4). 

# Create a new column 'Month'
df['Month'] = df['Order.Date'].dt.month

# Filter for Q4 (Months = 10, 11, 12)
q4_df = df[df['Month'].isin([10, 11, 12])]

# Show the first few rows
print(q4_df.head())
```

```python
#2.	Calculate the average profit for each city during that time. 
# Group by 'City' and calculate the mean of 'Profit'
average_profit_per_city = q4_df.groupby('City')['Profit'].mean().reset_index()

# Sort cities by Average Profit (optional, to make bar plot prettier)
average_profit_per_city = average_profit_per_city.sort_values(by='Profit', ascending=False)

# Show the result
print(average_profit_per_city.head())
```

```python
#3.	Display the results using a bar plot and label the axes clearly (e.g., “Average Profit (December) by City”). 

# Select top 10 cities by average profit
top_cities = average_profit_per_city.sort_values('Profit', ascending=False).head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.bar(top_cities['City'], top_cities['Profit'], color='skyblue')

plt.title('Top 10 Cities - Average Profit (December)', fontsize=18)
plt.xlabel('City', fontsize=14)
plt.ylabel('Average Profit', fontsize=14)

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
